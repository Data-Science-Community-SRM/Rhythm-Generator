```python
import seaborn as sns
```

Correllation matrices using different methods

```python
corr=rhyth_gen.corr(method="pearson")
plt.subplots(figsize=(10,10))
sns.heatmap(corr,cmap="BuPu",annot=True)
```

```python
corr=rhyth_gen.corr(method="kendall")
plt.subplots(figsize=(10,10))
sns.heatmap(corr,cmap="BuPu",annot=True)
```

Checking for null values in the dataset

```python
rhyth_gen.isnull().sum()
```

Droping duplicate values

```python
rhyth_gen=rhyth_gen.drop_duplicates()
```

Transposing the columns and rows of the dataset

```python
rhyth_gen.head(10).T
```

```python
from matplotlib.pyplot import show
total = float(len(rhyth_gen)) 
sns.set(style="darkgrid")

ax = sns.countplot(x="pitch_range",data=rhyth_gen)

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}'.format(height/total),
            ha="center")
show()
```

A scatter plot between minimum and maximum velocity to show how one is affected by another

```python
plt.scatter(rhyth_gen['min_velocity'],rhyth_gen['max_velocity'])
plt.xlabel("Minimun Velocity")
plt.ylabel("Maximum Velocity")
plt.title('Min vs Max Velocity')
plt.show()
```

Analysing what kind of beats per minute do the track roles have

```python
plt.bar(rhyth_gen['track_role'],rhyth_gen['bpm'],color='purple')
plt.xlabel("Track Roles")
plt.ylabel("Beats Per Minute")
plt.show()
```

```python
plt.bar(rhyth_gen['genre'],rhyth_gen['bpm'],color='green')
plt.xlabel("Genres")
plt.ylabel("Beats Per Minute")
plt.title("Cinematic Vs Newage")
plt.show()
```

```python
plt.bar(rhyth_gen['track_role'],rhyth_gen['num_measures'],color='darkblue')
plt.xlabel("Track Roles")
plt.ylabel(" Num Measures")
plt.show()
```
